```python
import pymongo
import os

MONGO_USERNAME = os.getenv("MONGO_INITDB_ROOT_USERNAME")
MONGO_PASSWORD = os.getenv("MONGO_INITDB_ROOT_PASSWORD")

mongo = pymongo.MongoClient(f"mongodb://{MONGO_USERNAME}:{MONGO_PASSWORD}@mongo:27017")
mongo.twitter.sample.find_one()
```

```python
import pandas as pd

pd.set_option('display.max_columns', 20)

tweets = list(mongo.twitter.sample.find(limit=500))
pdf = pd.DataFrame(tweets)
pdf.head(5)
```

```python
from pandas.io.json import json_normalize
import json

user_pdf = json_normalize(pdf["user"]).add_prefix("user_")
pdf = pdf.join(user_pdf).loc[:, ["user_name", "user_followers_count", "text", "keyword"]]
pdf.head(10)
```

```python
from pyspark.sql import SparkSession, SQLContext
from pyspark.context import SparkContext
from pyspark.sql.types import *

schema = StructType([
    StructField('user_name', StringType(), False),
    StructField('user_followers_count', IntegerType(), False),
    StructField('text', StringType(), False),
    StructField('keyword', StringType(), False)
])

sc = SparkContext.getOrCreate()
sdf = SQLContext(sc).createDataFrame(pdf, schema=schema)
sdf.show()
```

```python
sdf.createOrReplaceTempView("tweets")

spark = SparkSession.builder.getOrCreate()
query_result = spark.sql("SELECT user_name, text FROM tweets WHERE keyword = 'Python' LIMIT 5")
query_result.show()
```
